## What's Pandas for?
The Pandas package is the most important tool for Data Scientists working in Python today. Pandas is the backbone of most data projects.

Pandas allows to:
- Calculate statistics and answer questions about the data, like:
    - What's the average, median, max, or min of each column?
    - Does column A correlate with column B?
    - What does the distribution of data in column C look like?
- Clean the data by doing things like removing missing values and filtering rows or columns by some criteria.
- Visualize the data with help from Matplotlib. Plot bars, lines, histograms,bubbles, and more.
- Store the cleaned, transformed data back into a CSV, other file or database.

### Querying
A spreadsheet is useless if you can't dice/sort/etc.

```python
# rating less than 3
books_df.rating < 3
```

```python
import numpy as np
np.sum(books_df.rating < 3)
```

Why did that work?

```python
print(1*True, 1*False)
```

So we ought to be able to do this

```python
print(np.sum(books_df.rating < 3) / books_df.shape[0])
# print("%.5f" % (np.sum(books_df.rating < 3)/books_df.shape[0]))
```

Notice that you could just find the average since the Trues map to 1s.

```python
np.mean(books_df.rating < 3.0)
```

Or directly in Pandas, since **books_df.rating < 3** is a pandas Series.

```python
print((books_df.rating < 3).mean())
# print((books_df["rating"] < 3).mean())
```

### Filtering
Here are two ways to get a filtered dataframe:

```python
books_df.query('rating > 4.5')
```

Here we create a mask and use it to 'index' into the dataframe to get the rows we want.

```python
books_df[books_df.year < 0]
```

```python
# Combining the last two queries
books_df[(books_df.year < 0) & (books_df.rating > 4)]
```

### Cleaning

Let's try the following:

```python
books_df['rating_count'] = books_df.rating_count.astype(int)
books_df['review_count'] = books_df.rating_count.astype(int)
books_df['year'] = books_df.rating_count.astype(int)
```

Oops! We got an error! This means something is missing.

```python
books_df[books_df.year.isnull()]
```

```python
books_df = books_df[books_df.year.notnull()]
books_df.shape
```

We removed those 7 rows. Let's try again:

```python
books_df['rating_count'] = books_df.rating_count.astype(int)
books_df['review_count'] = books_df.rating_count.astype(int)
books_df['year'] = books_df.rating_count.astype(int)
```

## Visualizing

```python
books_df.rating.hist();
```

```python
import seaborn as sns
sns.set_context("notebook")

# meanrat = books_df["rating"].mean()
meanrat = books_df.rating.mean()

# you can get means and medians in different ways
print(meanrat)
print(np.mean(books_df.rating))
print(books_df.median())
```

```python
import matplotlib
import matplotlib.pyplot as plt

with sns.axes_style("whitegrid"):
    books_df.rating.hist(bins=30, alpha=0.4);
    plt.axvline(meanrat, 0, 0.75, color='r', label='Mean')
    plt.xlabel("average rating of book")
    plt.ylabel("Counts")
    plt.title("Ratings Histogram")
    plt.legend()
    # sns.despine()
```

One can see the sparseness of review counts. This will be important when we learn about recommendations: we'll have to regularize our models to deal with it.

```python
books_df.review_count.hist(bins=np.arange(0, 40000, 400));
```

The structure may be easier to see if we rescale the x-axis to be logarithmic.

```python
books_df.review_count.hist(bins=100)
plt.xscale("log")
```

```python
plt.scatter(books_df.year, books_df.rating, lw=0, alpha=.9)
plt.xlim([1900,2010])
plt.xlabel("Year")
plt.ylabel("Rating");
```
